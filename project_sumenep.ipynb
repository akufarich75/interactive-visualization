```python
#menghapus kolom yang banyak unique dan missing values
data.drop(['NO. KK','KETERANGAN BLOK II','WANITA USIA HAMIL','KETERANGAN BLOK II',
           'KETERANGAN BLOK IV','KETERANGAN KB','KETERANGAN BLOK V'], axis=1, inplace=True)
```

```python
data2.drop(['NO. KK','ART MINUM OBAT GANGGUAN JIWA BERAT TERATUR','KETERANGAN BLOK II','WANITA USIA HAMIL','KETERANGAN BLOK II',
           'KETERANGAN BLOK IV','KETERANGAN KB','KETERANGAN BLOK V'], axis=1, inplace=True)
```

```python
#CLEANING DATA
data.dropna(subset=['NAMA KK'], inplace=True)
data.dropna(subset=['NO. URUT KELUARGA'], inplace=True)
data.dropna(subset=['JENIS SUMBER AIR TERLINDUNG'], inplace=True)
data.dropna(subset=['JENIS JAMBAN SANITER'], inplace=True)
data.dropna(subset=['ART MINUM OBAT GANGGUAN JIWA BERAT TERATUR'], inplace=True)
data.dropna(subset=['NIK'], inplace=True)
```

```python
missing_data_object_column2= [i for i in data2.columns if data2[i].isnull().sum()>=0 and data2[i].dtype=='O']
missing_categorical_data2 = missing_data_object_column2
missing_numerical_column2 =[i for i in data2.columns if data2[i].isnull().sum()>=0 and data2[i].dtype=='int' and data2[i].dtype=='float']
missing_numerical_data2 = missing_numerical_column2
def removing_missing_value(data2): #menghapus missing value untuk numerikal dan kategorikal data sesuai dengan keadaan skew
    for col in missing_categorical_data2:
      data2[col].fillna(method='ffill',inplace=True)
    for col in missing_numerical_data2:
      data2[col].dropna(inplace=True)
removing_missing_value(data2)
```
