```python
def animated_graph(data4, column_name):
    fig = px.choropleth(data4.sort_values('year'), locations="Country name", color=column_name, 
                    locationmode='country names',
                    animation_frame='year',
                 color_continuous_midpoint=0.8,
                    color_continuous_scale=px.colors.sequential.dense,
#                         width=800, height=400
                   )

    fig.update_layout(title=column_name+' Comparison by country',autosize=True)
    fig.show() 
```

```python


animated_graph(data4,"Perceptions of corruption")

```

```python
animated_graph(data4,"Freedom to make life choices")
```

```python
temp = data4[['Log GDP per capita','Life Ladder','year','Country name',
           'Healthy life expectancy at birth','Social support',
           'Freedom to make life choices','Perceptions of corruption','Generosity']]
is_NaN = temp.isnull()
row_has_NaN = is_NaN.any(axis=1)
rows_with_NaN = temp[row_has_NaN]
# print(len(rows_with_NaN))
# temp = temp.dropna(how='any',axis=0) 
temp = temp[~temp.year.isin([2006,2005,2007,2008])]
temp = temp.sort_values('year')
```

```python
all_countries = data4['Country name'].value_counts().reset_index()['index'].tolist()
all_countries_pop = data2['Country Name'].value_counts().reset_index()['index'].tolist()
del_cou = []
for x in all_countries:
    if x not in all_countries_pop:
        del_cou.append(x)
```

```python
pop_df = temp.copy()
pop_df = pop_df[~pop_df['Country name'].isin(del_cou)]
pop_df = pop_df[~pop_df.year.isin([2006,2005,2007,2018,2019,2020,2021])]
pop_dict = {x:{} for x in range(2008,2018)}
for i in range(len(data2)):
    if(data2['Year'][i] in range(2008,2018)):
        pop_dict[data2['Year'][i]][data2['Country Name'][i]] = data2['Count'][i]

population = []
for i in pop_df.index:
    population.append(pop_dict[pop_df['year'][i]][pop_df['Country name'][i]])
pop_df['population'] = population
```

```python


px.scatter(pop_df, x="Log GDP per capita", y="Life Ladder", animation_frame="year", 
           animation_group="Country name",size='population',color='Country name'
           , hover_name="Country name", size_max=60,
           color_discrete_sequence=px.colors.qualitative.T10)

```
